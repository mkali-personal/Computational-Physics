```python
import numpy as np
from matplotlib import pyplot as plt


def find_zero(F, dF, X, epsF, epsX, no_flipping=False):
    """
    Find the root of the multivariable vector function f.
    Args:
        f (`function`): The function for which to find a root. Accepts `x` (a vector) and returns a vector.
        df (`function`): The differential function of `f`. Accepts `x` and returns the Jacobian matrix at `x`.
        x (`np.array`): The initial guess for a root.
        eps (`float`): .
    Returns:
        x (`np.array`): The x value of the root of the function in the range [a,b].
    Raises:
        ValueError:
    """
    if np.linalg.norm(F(X)) < epsF:
        # Initial guess is close enough
        return X
    
    X = run_step(F, dF, X, no_flipping)
    
    while np.linalg.norm(F(X)) >= epsF:
        X = run_step(F, dF, X, no_flipping)

    return X

def run_step(F, dF, X_guess, no_flipping=False):
    delta = newton_step(F, dF, X_guess)
    print('Guess:', X_guess, 'Delta:', delta)
    
    if no_flipping:
        alpha = scale_to_prevent_flipping(X_guess, delta)
        delta = alpha * delta
        
    delta = line_search(F, X_guess, delta)
    return X_guess + delta
    

def scale_to_prevent_flipping(X, delta):
    Xd = X + delta - np.roll(X + delta, 1)
    Xd[0] = X[0] + delta[0]
    if any(Xd < 0):
        X_1d = X - np.roll(X, 1)
        X_1d[0] = X[0]
    
        deltad = delta - np.roll(delta, 1)
        deltad[0] = delta[0] # Set boundary (delta1 - delta0 = delta1)
    
        alpha_n = - X_1d / (deltad)
        return 0.9 * min(alpha_n[Xd < 0])
    return 1

def is_delta():
    pass

def newton_step(F, dF, X):
    J = dF(X)
    delta = - np.linalg.solve(J, F(X))
    
    return delta

def fmin(F, X):
    return 0.5 * sum(F(X)**2)

def line_search(F, X_old, delta):
    plt.figure(figsize=(12, 5))
    x = np.geomspace(1e-10, 1, num=100)
    y = np.zeros(x.size)
    for i in range(y.size):
        y[i] = fmin(F, X_old+x[i]*delta)
    plt.plot(x, y, '.')
    plt.grid()
    plt.show()

#     scale_factor = 1e16
#     if sum(delta**2) > scale_factor:
#         delta *= scale_factor / sum(delta**2)
    
    # Based majorly on the C implementation found in "Numerical Recipes in C, 2nd Edition" by W. H. Press et. al., p.385-386
    g_0 = fmin(F, X_old)
    dg_0 = - 2 * g_0
    
    l_1 = 1
    
    while True:
        X = X_old + l_1*delta
        g = fmin(F, X)
        print('\t\tl = %12.4e\tg(l) = %20.16f' % (l_1, g))
        
        if (g <= g_0 + 1e-4 * l_1 * dg_0):
            # Sufficient function decrease
            return l_1*delta
        
        # Find suitable lambda.
        if l_1 == 1:
            # First iteration, approximate g as a quadratic polynomial
            import pdb; pdb.set_trace()
            l = - (dg_0) / (2 * (g - g_0 - dg_0))
        
        else:
            a = (1 / (l_1 - l_2)) * ((1 / (l_1**2)) * (g - dg_0 * l_1 - g_0) + (-1 / (l_2**2)) * (g_2 - dg_0 * l_2 - g_0))
            b = (1 / (l_1 - l_2)) * ((-l_2 / (l_1**2)) * (g - dg_0 * l_1 - g_0) + (l_1 / (l_2**2)) * (g_2 - dg_0 * l_2 - g_0))
            
            if a == 0:
                l = - dg_0 / (2 * b)
            else:
                d = b**2 - 3 * a * dg_0 # the discriminant
                if d < 0:
                    l *= 0.5
                elif b <= 0:
                    l = (-b + np.sqrt(d)) / (3 * a)
                else:
                    l = - dg_0 / (b + np.sqrt(d))
            l = min(l, 0.5 * l_1)
    
        l_2, l_1 = l_1, max(l, 0.1 * l_1)
        
        g_2 = g

```

# White Dwarf Star


Solve multivariable vector function to find the structure of a white dwarf star.


$$f_{i}=\frac{2\kappa3^{\gamma}}{\left(4\pi\right)^{\gamma-1}Gm_{i}\left(m_{i+1}-m_{i-1}\right)}r_{i}^{4}\left[\left(\frac{m_{i+1}-m_{i}}{r_{i+1}^{3}-r_{i}^{3}}\right)^{\gamma}-\left(\frac{m_{i}-m_{i-1}}{r_{i}^{3}-r_{i-1}^{3}}\right)^{\gamma}\right]+1$$

$$ \frac{\partial f_{i}}{\partial r_{i}}=\frac{2\kappa3^{\gamma}}{\left(4\pi\right)^{\gamma-1}Gm_{i}\left(m_{i+1}-m_{i-1}\right)}r_{i}^{3}\left[4\left[\left(\frac{m_{i+1}-m_{i}}{r_{i+1}^{3}-r_{i}^{3}}\right)^{\gamma}-\left(\frac{m_{i}-m_{i-1}}{r_{i}^{3}-r_{i-1}^{3}}\right)^{\gamma}\right]+3r_{i}^{3}\gamma\left[\frac{\left(m_{i+1}-m_{i}\right)^{\gamma}}{\left(r_{i+1}^{3}-r_{i}^{3}\right)^{\gamma+1}}+\frac{\left(m_{i}-m_{i-1}\right)^{\gamma}}{\left(r_{i}^{3}-r_{i-1}^{3}\right)^{\gamma+1}}\right]\right] $$

$$ \frac{\partial f_{i}}{\partial r_{i-1}}=-\frac{2\kappa3^{\gamma}}{\left(4\pi\right)^{\gamma-1}Gm_{i}\left(m_{i+1}-m_{i-1}\right)}3\gamma r_{i}^{4}r_{i-1}^{2}\frac{\left(m_{i}-m_{i-1}\right)^{\gamma}}{\left(r_{i}^{3}-r_{i-1}^{3}\right)^{\gamma+1}} $$

$$ \frac{\partial f_{i}}{\partial r_{i+1}}=-\frac{2\kappa3^{\gamma}}{\left(4\pi\right)^{\gamma-1}Gm_{i}\left(m_{i+1}-m_{i-1}\right)}3\gamma r_{i}^{4}r_{i+1}^{2}\frac{\left(m_{i+1}-m_{i}\right)^{\gamma}}{\left(r_{i+1}^{3}-r_{i}^{3}\right)^{\gamma+1}} $$

```python
UNITS = 'SUN'

# Gravitational Constant
G_CGS = 6.674 * 1e-8 # cm^3 * g^-1 * s^-2
G_SUN = 6.674 * 1e-8 * (7 * 1e10)**-3 * 2 * 1e33 # R^3 * M^-1 * s^-2

# kappa for gamma = 5 / 3
GAMMA_NOT_RELATIVISTIC = 5 / 3
KAPPA_NOT_RELATIVISTIC_CGS = 3.15 * 10**12 # cm^4 * g^-(2/3) * s^-2
KAPPA_NOT_RELATIVISTIC_SUN = 3.15 * 10**12 * (7 * 1e10)**-4 * (2 * 1e33)**(2/3) # R^4 * M^-(2/3) * s^-2

# kappa for gamma = 4 / 3
GAMMA_RELATIVISTIC = 4 / 3
KAPPA_RELATIVISTIC_CGS = 4.9 * 10**14 # cm * g^-(1/3) * s^-2
KAPPA_RELATIVISTIC_SUN = 4.9 * 10**14 * (7 * 1e10)**-1 * (2 * 1e33)**(1/3) # R * M^-(1/3) * s^-2

if UNITS.upper() == 'CGS':
    G = G_CGS
    KAPPA_NOT_RELATIVISTIC = KAPPA_NOT_RELATIVISTIC_CGS
    KAPPA_RELATIVISTIC = KAPPA_RELATIVISTIC_CGS
elif UNITS.upper() == 'SUN':
    G = G_SUN
    KAPPA_NOT_RELATIVISTIC = KAPPA_NOT_RELATIVISTIC_SUN
    KAPPA_RELATIVISTIC = KAPPA_RELATIVISTIC_SUN
else:
    raise Exception()
```

```python
def generate_F_and_dF(m, r, gamma, kappa):
    ## Notational convention for F and dF:
    # a    is a_{i}
    # a_1  is a_{i-1}
    # a__1 is a_{i+1}
    
    m_1 = np.roll(m, 1)
    m_1[0] = 0 # Set the boundary condition on m0
    m__1 = np.roll(m, -1)
    m__1[-1] = m[-1] # Set the boundary condition on m(n+1)        
    
    md_1 = m - m_1 # m(i) - m(i-1)
    md__1 = m__1 - m # m(i+1) - m(i)
    
    layer_mass = m - m_1

    star_constant = (2 * kappa * 3**gamma) / ((4 * np.pi)**(gamma-1) * G * m * (m__1 - m_1))

    def F(r):
        #print('Calculating F')
        r_1 = np.roll(r, 1)
        r_1[0] = 0 # Set boundary condition on r0
        
        rd_1 = r**3 - r_1**3
        
        a = (md_1 / rd_1)**gamma
        b = np.roll(a, -1)
        b[-1] = 0
        
        #print('r:', r)
        #print('F(r):', star_constant * r**4 * (b - a) + 1)
        
        return star_constant * r**4 * (b - a) + 1
        
    def dF(r):
        #print('Calculating dF')
        r_1 = np.roll(r, 1)
        r_1[0] = 0 # Set boundary condition on r0
        r__1 = np.roll(r, -1)
        r__1[-1] = r[-1]+1 # Set boundary condition on r(n+1) to R+1. This is so (r+1)^3 - r^3 doesn't equal 0. In the actual calculations it doesn't matter since when dividing by (r+1)^3 - r^3, we are always dividing 0.
        
        dFdr = star_constant * r**3 * ( 4 * ((md__1 / (r__1**3 - r**3))**gamma - (md_1 / (r**3 - r_1**3))**gamma) + 3 * gamma * r**3 * (md__1**gamma / (r__1**3 - r**3)**(gamma+1) + md_1**gamma / (r**3 - r_1**3)**(gamma+1)) )
        
        dFdr_1 =  - star_constant * 3 * gamma * r**4 * r_1**2  * (md_1**gamma / (r**3 - r_1**3)**(gamma+1))
        dFdr__1 = - star_constant * 3 * gamma * r**4 * r__1**2 * (md__1**gamma / (r__1**3 - r**3)**(gamma+1))
        
        #print('r:', r)
        #print('dFr:')
        #print(np.diag(dFdr_1, -1)[1:, 1:] + np.diag(dFdr) + np.diag(dFdr__1, 1)[:-1, :-1])
        
        return np.diag(dFdr_1, -1)[1:, 1:] + np.diag(dFdr) + np.diag(dFdr__1, 1)[:-1, :-1]
    
    return F, dF
```
